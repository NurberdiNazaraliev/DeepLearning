```python
def neural_network(input, weights):
    pred= 0 
    for i in range(len(input)):
        pred += input[i]*weights[i]
    return pred


def scalar_vector(scalar, vector):
    out= [0,0,0]
    for i in range(len(vector)):
        out[i]= scalar*vector[i]

    return out
```

```python
y_real = 1.5
alpha = 0.01
weights = [0.1, 0.2, 0]
input = [8.5, 0.65, 1.2]

for iter in range(15):
    pred = neural_network(input,weights)
    error = (pred - y_real) ** 2
    delta =  pred - y_real
    
    weight_deltas = scalar_vector(delta, input)
    
    # comments
    print("Iteration: " + str(iter+1))
    print("Pred: " + str(pred))
    print("Error: " + str(error))
    print("Delta: " + str(delta))
    print("Weights: " + str(weights))
    print("Weight_Deltas: ")
    print(str(weight_deltas))
    print()
    
    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]
```

Congratulations! We move on to the first real-world dataset.

Its called the Modified National Institute of Standards and Technology (MNIST) dataset,
and it consists of digits that high school students and employees of the US Census Bureau
handwrote some years ago. The interesting bit is that these handwritten digits are black-andwhite
images of peoples handwriting. Accompanying each digit image is the actual number
they were writing (09). For the last few decades, people
have been using this dataset to train neural networks to read
human handwriting, and today, youre going to do the same.

Each image is only 784 pixels (28 x 28). Given that you have
784 pixels as input and 10 possible labels as output, you
can imagine the shape of the neural network: each training
example contains 784 values (one for each pixel), so the
neural network must have 784 input values. Pretty simple,
eh? You adjust the number of input nodes to reflect how
many datapoints are in each training example. You want to
predict 10 probabilities: one for each digit. Given an input drawing, the neural network
will produce these 10 probabilities, telling you which digit is most likely to be what
was drawn.
