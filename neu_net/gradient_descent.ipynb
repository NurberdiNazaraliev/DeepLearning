## Learning using the hot and cold method

Hot and cold learning means wiggling the weights to see which direction reduces the error the most, moving the weights in that direction, and repeating until the error gets to 0.


````python
# How to calculate error
error = (pred - true) ** 2
print(error)
````

````python
# Making a prediction with a HIGHER weight and evaluating error
step = 0.01
pred = neural_network(input, weight+ step)
error = (pred - true) ** 2
print(error)
````

````python
# Making a prediction with a LOWER weight and evaluating error
step = 0.01

pred = neural_network(input, weight- step)
error = (pred - true) ** 2
print(error)
````

### Hot and cold learning

````python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.01

for iteration in range(150):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    
    print("Error:" + str(error) + " Prediction:" + str(prediction))

    # trying to predict using the higher weight
    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2

    # with lower weight
    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount

    if(down_error > up_error):
        weight = weight + step_amount
````

````python
# what if set step_amount to 10?

weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 10

for iteration in range(150):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    
    print("Error:" + str(error) + " Prediction:" + str(prediction))

    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2

    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount

    if(down_error > up_error):
        weight = weight + step_amount
````

## Calculating both direction and amount from error

````python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    
    # gradient descent
    direction_and_amount = (pred - goal_pred) * input
    
    weight = weight - direction_and_amount
    print("Error:" + str(error) + " Prediction:" + str(pred))
````

**What is direction_and_amount?**

`direction_and_amount` represents how you want to change weight. The first part is pure error, which equals `(pred - goal_pred)`. The second part is the multiplication by the `input` that performs scaling, negative reversal, and stopping, modifying the pure error so its ready to update `weight`.

**What is the pure error?**

The pure error is `(pred - goal_pred)`, which indicates the raw direction and amount you missed. If this is a positive number, you predicted too high, and vice versa. If this is a big number, you missed by a big amount, and so on.

**What are scaling, negative reversal, and stopping?**

These three attributes have the combined effect of translating the pure error into the absolute amount you want to change `weight`. They do so by addressing three major edge cases where the pure error isnt sufficient to make a good modification to `weight`.

## Introducing alpha
#### Its the simplest way to prevent overcorrecting weight updates

The solution is to multiply the weight update by a fraction to make it smaller. In most cases, this involves multiplying the weight update
by a single real-valued number between 0 and 1, known as alpha. Note: this has no effect on the core issue, which is that the input is larger. It will also reduce the weight updates for inputs that arent too large.

Finding the appropriate alpha, even for state-of-the-art neural networks, is often done by guessing. You watch the error over time. If it starts diverging (going up), then the alpha is too high, and you decrease it. If learning is happening too slowly, then the alpha
is too low, and you increase it.

````python
weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1

for iteration in range(40):
    pred= input*weight
    error= (pred-goal_pred)**2
    
    delta_weight= (pred-goal_pred)*input
    weight=weight- (alpha* delta_weight)

    
    print("Error:" + str(error) + " Prediction:" + str(pred))
````
