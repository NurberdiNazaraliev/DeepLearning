```python
import numpy as np
def neural_network(input, weights):
    pred= 0 
    for i in range(len(input)):
        pred += input[i]*weights[i]
    return pred


def scalar_vector(scalar, vector):
    out= [0,0,0,0,0]
    for i in range(len(vector)):
        out[i]= scalar*vector[i]

    return out
```

## 4. Calculating the Derivative of the Function using the TensorFlow (2 points).

You are given a function `y`, find the derivative at a given point `x`.

$ y = 5x^3 + x^2 - 0.5x +6$ 

Calculate the derivative of the function at $ x = 3 $.

```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.Variable(initial_value = 3.0)
with tf.GradientTape() as tape:
        y = 5*x**3 + x**2 - 0.5*x +6
gradient = tape.gradient(y, x) # gradient(y, x)
print(gradient)
         
```
